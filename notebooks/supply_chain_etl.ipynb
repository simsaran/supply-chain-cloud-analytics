```python
# Cell 2: Load and explore the data
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv(
  'DataCoSupplyChainDatasetRefined.csv',
  encoding='latin-1'
)

# Show how many rows and columns
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

# Show all column names
print("\nColumn names:")
for col in df.columns:
    print("  -", col)

# Show first 5 rows
df.head()
```

```python
# Cell 3: Check data quality

# Show data types and missing values
print("=== Data Types ===")
print(df.dtypes)

print("\n=== Missing Values ===")
missing = df.isnull().sum()
print(missing[missing > 0])
```

```python
# Cell 4: Clean and transform the data

# 1. Drop columns we don't need
drop_cols = [
  'Customer Email',
  'Customer Password',
  'Product Image',
  'Order Zipcode',
  'Product Description',
  'Product Status'
]
df = df.drop(
  columns=drop_cols,
  errors='ignore'
)

# 2. Clean column names
#    "Order Date" → "order_date"
df.columns = [
  col.strip()
     .replace(' ', '_')
     .replace('(', '')
     .replace(')', '')
     .replace('/', '_')
     .lower()
  for col in df.columns
]

# 3. Fix date columns
df['order_date'] = pd.to_datetime(
  df['order_date_dateorders'],
  errors='coerce'
)
df['shipping_date'] = pd.to_datetime(
  df['shipping_date_dateorders'],
  errors='coerce'
)

# 4. Create useful new columns
df['delivery_delay_days'] = (
  df['days_for_shipping_real']
  - df['days_for_shipment_scheduled']
)

df['late_delivery_flag'] = (
  df['delivery_delay_days'] > 0
).astype(int)

# 5. Create date parts for analysis
df['order_year'] = df['order_date'].dt.year
df['order_month'] = df['order_date'].dt.month
df['order_month_name'] = (
  df['order_date'].dt.month_name()
)
df['order_day_of_week'] = (
  df['order_date'].dt.day_name()
)

print("✅ Cleaning done!")
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
print(df.columns.tolist())
```

```python
# Cell 5: Create dimension tables

# dim_product — unique products
dim_product = df[[
  'product_card_id',
  'product_name',
  'category_name',
  'department_name',
  'product_price'
]].drop_duplicates().reset_index(drop=True)

print("dim_product:", dim_product.shape)

# dim_customer — unique customers
dim_customer = df[[
  'customer_id',
  'customer_fname',
  'customer_lname',
  'customer_segment',
  'customer_city',
  'customer_state',
  'customer_country'
]].drop_duplicates().reset_index(drop=True)

print("dim_customer:", dim_customer.shape)

# dim_region — unique regions
dim_region = df[[
  'market',
  'order_region',
  'order_country',
  'order_city'
]].drop_duplicates().reset_index(drop=True)
dim_region['region_id'] = range(
  1, len(dim_region) + 1
)

print("dim_region:", dim_region.shape)

# fact_orders — the main data
fact_orders = df[[
  'order_id',
  'order_item_id',
  'product_card_id',
  'customer_id',
  'order_date',
  'order_year',
  'order_month',
  'order_month_name',
  'order_day_of_week',
  'market',
  'order_region',
  'order_country',
  'order_city',
  'type',
  'days_for_shipping_real',
  'days_for_shipment_scheduled',
  'delivery_delay_days',
  'late_delivery_flag',
  'late_delivery_risk',
  'order_item_quantity',
  'sales',
  'order_profit_per_order',
  'benefit_per_order',
  'order_item_discount',
  'order_item_discount_rate',
  'order_item_profit_ratio',
  'shipping_mode'
]].copy()

print("fact_orders:", fact_orders.shape)
print("\n✅ Star schema created!")
```
